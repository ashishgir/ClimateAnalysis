```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('GlobalTemperatures.csv')
```

```python
df
```

```python
df = df[['dt', 'LandAverageTemperature', 'LandMaxTemperature', 'LandMinTemperature', 'LandAndOceanAverageTemperature']]
```

```python
df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
df['dt'] = pd.to_datetime(df['dt'])
```

```python
df['date_delta'] = (df['dt'] - df['dt'].min())  / np.timedelta64(1,'D')
df = df[['date_delta', 'LandAverageTemperature', 'LandMaxTemperature', 'LandMinTemperature', 'LandAndOceanAverageTemperature']]
```

```python
X_train = df[['date_delta', 'LandMaxTemperature', 'LandMinTemperature', 'LandAndOceanAverageTemperature']]
y_train = df['LandAverageTemperature']
```

```python
from sklearn.linear_model import LinearRegression

lineReg = LinearRegression()
lineReg.fit(X_train, y_train)
```
